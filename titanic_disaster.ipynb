Titanic_disaster_survival prediction


```python
import pandas as pd
```

Loading the Data

```python
train_df=pd.read_csv('datasets/train.csv')
test_df=pd.read_csv('datasets/test.csv')
```

```python
train_df
```

```python
test_df
```

Separate Features and Target

```python
X_train=train_df.drop('Survived',axis=1)
y_train=train_df['Survived']
X_test=test_df.copy()
```

```python
X_train.info()
```

```python
X_train.isnull().sum()
```

Handling Missing Values

```python
from sklearn.impute import SimpleImputer
age_imputer=SimpleImputer(strategy='mean')

X_train['Age']=age_imputer.fit_transform(X_train[['Age']])
```

```python
X_test['Age']=age_imputer.transform(X_test[['Age']])
```

```python
X_train.info()
```

Encoding Categorical Features

```python
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(handle_unknown='ignore',sparse_output=False)

encoded_train=encoder.fit_transform(X_train[['Sex']])
encoded_test=encoder.transform(X_test[['Sex']])
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
numerical_columns=["Age","Fare","SibSp","Parch"]
X_train_numeric=X_train[numerical_columns]
X_test_numeric=X_test[numerical_columns]
X_train_scaled=scaler.fit_transform(X_train_numeric)
X_test_scaled=scaler.transform(X_test_numeric)
```

Drop High_missing Columns

```python
X_train.drop('Cabin',axis=1,inplace=True)
X_test.drop('Cabin',axis=1,inplace=True)
```

handling missing value(Embarked)

```python
embarked_imputer=SimpleImputer(strategy='most_frequent')
X_train['Embarked']=embarked_imputer.fit_transform(X_train[['Embarked']]).ravel()
```

```python
X_test['Embarked']=embarked_imputer.transform(X_test[['Embarked']]).ravel()
```

Categorical Feature(Embarked)

```python
from sklearn.preprocessing import OneHotEncoder
embark_encoder=OneHotEncoder(handle_unknown='ignore',sparse_output=False)
X_train_emb=embark_encoder.fit_transform(X_train[['Embarked']])
X_test_emb=embark_encoder.transform(X_test[['Embarked']])
```

combining all features

```python
import numpy as np

X_train_final=np.hstack([X_train_scaled,encoded_train,X_train_emb])
X_test_final=np.hstack([X_test_scaled,encoded_test,X_test_emb])
```

```python
X_train_final.shape
```

```python
X_test_final.shape
```

```python
X_train_final
```

Train-Validation Split

```python
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(X_train_final,y_train,test_size=0.2,random_state=42)
```

training model(Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(max_iter=1000)
classifier.fit(X_tr,y_tr)
```

Model Evaluation

Predictions

```python
y_pred=classifier.predict(X_val)
```

Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_val,y_pred)
```

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_val,y_pred)
```

Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_val,y_pred))
```

Training model(RandomForest)

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200,random_state=42)
rf.fit(X_tr,y_tr)
```

Evaluation

Prediction

```python
rf_pred=rf.predict(X_val)
```

Accuracy Score

```python
accuracy_score(y_val,rf_pred)
```

Creating submission file(comparing with test dataset)

```python
final_model=rf
test_predictions=final_model.predict(X_test_final)
submission = pd.DataFrame({'PassengerId':test_df['PassengerId'],'Survived':test_predictions})

submission.to_csv('submission.csv',index=False)
```
